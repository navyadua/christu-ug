####Mean Median And Mode

```
mean=df['arrival_date_year'].mean()
mean
```

```
median=df['arrival_date_year'].median()
median
```

```
mode = df['arrival_date_year'].mode()
mode = mode[0]
mode
```

####Vector Sum and Product

```
df['sum_of_total_nights'] = df['stays_in_week_nights'] + df['stays_in_weekend_nights']
df['sum_of_total_nights']
```

```
df['Total_Rate'] = df['sum_of_total_nights'] * df['adr']
df['Total_Rate']
```

####Range, Quartile Deviation, Standard Deviation and Coefficient Of Variation on Grouped Data

```
df2 = df.groupby(by = ['country'])
df2.count()
```

```
max = df2['adr'].max()
min = df2['adr'].min()
range = max - min
range
```

```
q1=df2['adr'].quantile(0.25)
q3=df2['adr'].quantile(0.75)
iqr = q3 - q1
quar_dev=iqr/2
quar_dev
```

```
std = df2['adr'].std()
std
```

```
cov=df2['adr'].std()/df2['adr'].mean()
cov
```

###Result

#### Measures of Central Tendency

Measures of Central Tendency have been calculated for arrival date year

```
moc = pd.Series({'Mean ' : mean , 'Median' : median , 'Mode' : mode })
moc
```

Since Mean , Median , Mode are mostly 2016,
Hence we can interpret that most hotels were booked around the year 2016

#### Vector Sum and Product

Vector Sum has been calculated for : Lead Time and Days in Waiting list

```
vector_sum = pd.DataFrame({'Stay in Week Nights' : df['stays_in_week_nights'] ,
                           'Stay in Weekend Nights' : df['stays_in_weekend_nights'] ,
                           'Sum of Total Nights': df['sum_of_total_nights']})
vector_sum
```

Vector Product has been calculated for Total Night Stay and Average Daily Rate

```
vector_product = pd.DataFrame({'Total Nights Stay' : df['sum_of_total_nights'] ,
                               'Average Daily rate ' : df['adr'] ,
                               'Total Rate ' : df['Total_Rate']})
vector_product
```

#### Measures of Dispersion

These measures of dispersion have been calculated on Average Daily Rate and have been grouped on basis of countries

```
mod = pd.DataFrame({'Range' : range ,
                    'Quartile Deviation' : quar_dev ,
                    'Standard Deviation' : std ,
                    ' Coefficient of Variation ' :cov })
mod
```

From the measures of dispersion we may conclude that the data is quite scattered
