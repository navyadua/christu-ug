#PREPROCESSING

Import necessary libraries

```
import numpy as np
import pandas as pd
```

```
import warnings
warnings.filterwarnings('ignore')
```

Import Dataset

```
df = pd.read_csv('https://raw.githubusercontent.com/navyadua/data/main/hotel_booking_demand.csv',na_values = ['no info', '?'])
df
```

Counting number of null values

```
df.isna().sum()
```

Percentage of null values in each column

```
df.isna().sum()/len(df)*100
```

```
df.dropna(axis=0)
```

Company column haas too many null values, so we will delete the column so as to perform various operations

```
df.drop(columns = ['company'] , axis = 1 , inplace = True)
df.columns[df.isnull().any()]
```

```
null = df[df.isnull().any(axis=1)]
null
```

Dropping all rows with null values

```
df.dropna(axis = 0,inplace = True)
```

```
df.shape
```

```
df.isna().sum()
```

```
df.isna().sum()/len(df)*100
```

```
df.describe()
```

###Standardisation

```
from sklearn import preprocessing
df['adr_standardised']=preprocessing.scale(df.adr)
df
```

```
df.describe()
```

##Normalisation

```
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
array=np.array(df.adr).reshape(-1,1)
df['adr_normalised']=pd.DataFrame(scaler.fit_transform(array))
df
```

```
df.describe()
```
