#CROSS TABULATION & DESCRIPTIVE STATISTICS

```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/navyadua/data/main/hotel_booking_demand_perfect.csv')
df
```

```python
pd.crosstab(index=df['arrival_date_month'], columns=df['market_segment'])
```

```python
pd.crosstab(index=df['customer_type'],
            columns=df['deposit_type'])
```

```python
pd.crosstab(index=df['assigned_room_type'],
            columns=df['reservation_status'])
```
